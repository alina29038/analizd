```python
import pandas as pd
import numpy as np
```

импорт датасета

```python
df = pd.read_csv('bbc-news-data.csv', sep='\t')
df.head()
```

удаление знаков препинания и др. ненужных символов

```python
df = df.replace(r'[.,"\'-?:!;/<>()*\\]', '', regex=True)
df
```

приведение к нижнему регистру

```python
df['content'] = df['content'].str.lower()
df['title'] = df['title'].str.lower()
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
```

векторизация - деление предложения по словам

```python
for i in range(len(df['content'])):
  text = nltk.word_tokenize(df['content'][i])
  df['content'][i] = text

for i in range(len(df['title'])):
  text = nltk.word_tokenize(df['title'][i])
  df['title'][i] = text
```

удаление колонки "filename", за ненадобностью

```python
df =df.drop('filename', axis = 1)
df
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

```

удаление стоп слов

```python
for i in range(len(df['title'])):
  text = df['title'][i]
  df['title'][i]= [word for word in text if word not in stop_words]
for i in range(len(df['content'])):
  text = df['content'][i]
  df['content'][i]= [word for word in text if word not in stop_words] 
df
```

кодирование колонки "category"

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['category']= label_encoder.fit_transform(df['category'])
df
```

```python
nltk.download('wordnet')
```

лемматизация - преобразование слов в неопределенную форму 


```python
from nltk.tag.brill import Word
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for i in range(len(df['title'])):
  text = df['title'][i]
  df['title'][i]= [lemmatizer.lemmatize(word) for word in text]
for i in range(len(df['content'])):
  text = df['content'][i]
  df['content'][i]= [lemmatizer.lemmatize(word) for word in text]
df
```

колонку "content" преобразую в цифры

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['content'].astype('U'))
y = df['category']

```

логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(X, y)
```

```python
logisticRegr.score(X,y)
```

градиентный бустринг
