```python
from sklearn import datasets
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = datasets.load_boston()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df = df.drop(columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM', 'AGE','DIS', 'RAD', 'TAX', 'PTRATIO', 'B'])
df['PRICE'] = data.target
df.head()
```

```python
# import numpy as np
# from sklearn.linear_model import LinearRegression
y = df['PRICE']
X = df['LSTAT']
# reg = LinearRegression().fit(X, y)
# reg.score(X, y)
```

```python
Xn = len(X)
i = 0
Xsr = 0
for j in range(Xn):
  Xsr += df.iloc[i]['LSTAT']
  i+=1
Xsr = Xsr/Xn
print('Xsr: ',Xsr)

Yn = len(y)
i = 0
Ysr = 0
for j in range(Yn):
  Ysr += df.iloc[i]['PRICE']
  i+=1
Ysr = Ysr/Yn
print('Ysr: ',Ysr)
```

```python
Xsrkv = 0
i = 0
for i in range(Xn):
  Xsrkv += (df.iloc[i]['LSTAT'] - Xsr) ** 2
  i+=1
Xsrkv = Xsrkv/Xn
Xsrkv = Xsrkv ** 0.5
print('Xsrkv: ',Xsrkv)

Ysrkv = 0
i = 0
for i in range(Yn):
  Ysrkv += (df.iloc[i]['PRICE'] - Ysr) ** 2
  i+=1
Ysrkv = Ysrkv/Yn
Ysrkv = Ysrkv ** 0.5
print('Ysrkv: ',Ysrkv)
```

```python
i = 0
xysr = 0
for i in range(Xn):
  xysr += df.iloc[i]['LSTAT'] * df.iloc[i]['PRICE']
  i+=1
xysr= xysr/Xn
coefcorr = (xysr - Xsr * Ysr) / (Xsrkv * Ysrkv)
print('coefcorr' ,coefcorr)
```

```python
a = (coefcorr * Ysrkv) / Xsrkv
b = Ysr - a * Xsr
print('a: ',a)
print('b: ',b)
```
