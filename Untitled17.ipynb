импорт библиотек

```python
import warnings
import itertools
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

импорт датасета

```python
df = pd.read_csv('AirPassengers.csv', index_col='Month', parse_dates=['Month'])
df.head()
```

визуализация

```python
df.plot()
plt.show()
```

```python
import statsmodels.api as sm

fig = plt.figure(figsize=(100, 8))
decomposition = sm.tsa.seasonal_decompose(df)
decomposition.plot()
plt.show()
```

сама модель вот
параметры помню брала наобум, проэтомк не судите строго

```python
from statsmodels.tsa.arima.model import ARIMA
# Создать модель ARIMA
model = ARIMA(df, order=(1, 1, 1))

# Обучить модель на данных
model_fit = model.fit()

pred = model_fit.predict(start='1956-01-01', end='1960-12-01')
```

визуализация исходного и того что предсказала модель для сравнения и чтоб увидеть точность

```python
plt.figure(figsize=(20, 15))
plt.plot(df)
plt.plot(pred, color='red')
plt.show()
```
