```python
import pandas as pd
import numpy as np
```

```python
ptd1 = pd.read_csv('part_01.csv')
ptd2 = pd.read_csv('part_02.csv')
ptd3 = pd.read_csv('part_03.csv')
ptd4 = pd.read_csv('part_04.csv')
ptd5 = pd.read_csv('part_05.csv')
ptd6 = pd.read_csv('part_06.csv')
ptd7 = pd.read_csv('part_07.csv')
ptd8 = pd.read_csv('part_08.csv')
ptd9 = pd.read_csv('part_09.csv')
ptd10 = pd.read_csv('part_10.csv')
ptd11 = pd.read_csv('part_11.csv')
ptd12 = pd.read_csv('part_12.csv')

ptd1['ID'].unique
```

```python
gtd1 = pd.concat([ptd5, ptd12], ignore_index=True)
gtd2 = pd.concat([ptd6, ptd10], ignore_index=True)
gtd3 = pd.concat([ptd2, ptd11], ignore_index=True)
gtd4 = pd.concat([ptd1, ptd9], ignore_index=True)
gtd5 = pd.concat([ptd3, ptd8], ignore_index=True)
gtd6 = pd.concat([ptd4, ptd7], ignore_index=True)
```

```python
final_dataset = gtd1.merge(gtd2, left_on='ID', right_on='ID')
final_dataset = final_dataset.merge(gtd3, left_on='ID', right_on='ID')
final_dataset = final_dataset.merge(gtd4, left_on='ID', right_on='ID')
final_dataset = final_dataset.merge(gtd5, left_on='ID', right_on='ID')
final_dataset = final_dataset.merge(gtd6, left_on='ID', right_on='ID')
final_dataset = final_dataset.sort_values(by='ID')
final_dataset
```

```python
final_dataset = final_dataset.drop_duplicates(keep= False)
```

```python
final_dataset.sort_values(by='ID')
```
